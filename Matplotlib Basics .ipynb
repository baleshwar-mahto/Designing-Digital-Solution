## How to draw multiple lines in a same plot

```python
import numpy as np
```

```python
x = np.linspace(-np.pi,np.pi,256,endpoint = True) # endpoint = True implies that the end point of the series.i.e -pi and pi are also to be included
x
```

```python
S = np.sin(x)
C= np.cos(x)
```

```python
plt.plot(x,C)
plt.plot(x,S)
plt.show()
```

## How to add legends to the graph

```python
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')

plt.legend()
plt.show()
```

## How to increase figure size and resolution

```python
plt.figure(figsize =(8,3),dpi = 200)
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')

plt.legend()
plt.show()
```

### How to add grids to the graph

```python
plt.figure(figsize =(8,3),dpi = 200)
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')
plt.legend()

plt.grid(True)
plt.show()
```

### Setting axis limits

```python
plt.figure(figsize =(8,3),dpi = 200)
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')
plt.legend()
plt.grid(True)

plt.axis([0,5,0,1])
plt.show()
```

### Adding the axis labels

```python
plt.figure(figsize =(8,3),dpi = 200)
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')

plt.xlabel('This is X-axis')
plt.ylabel('This is Y-axis')

plt.show()
```

###  Give Title to your graph

```python
plt.figure(figsize =(8,3),dpi = 200)
plt.plot(x,C,label ='Cos')
plt.plot(x,S,label = 'Sin')

plt.xlabel('This is X-axis')
plt.ylabel('This is Y-axis')

plt.title('My first python plot')

plt.show()
```

### Interactive Graphs

```python
from ipywidgets import interact
```

```python
def interactive_plot(a,b,c,d,color,linestyle,linewidth):
    fig = plt.figure()
    axes = fig.add_subplot(111)
    x = np.linspace(-10,10,20)
    y = a*(x**2)+b*(x**2)+c*x+d
    title = "$f(x) = (%s)x^3 + (%s)x^2 + (%s)x + (%s)$" % (a,b,c,d)
    axes.plot(x,y,color = color,linestyle=linestyle,linewidth=linewidth)
    axes.set_title(title)
    axes.grid()
    plt.show()
    
```

```python
i = interact(interactive_plot,
           a = (-10,10),
           b = (-10,10),
           c = (-10,10),
           d = (-10,10),
           color = ['red','blue','green'],
           linestyle = ['solid','dashed'],
           linewidth = (1,5)
           )
```
